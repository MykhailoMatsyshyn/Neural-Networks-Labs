```python
import requests
import json
import sseclient  # Встановіть: pip install sseclient-py
from openai import OpenAI

# Ініціалізація клієнта OpenAI (для консистентності, не використовується в запиті)
client = OpenAI(base_url="http://192.168.0.105:1234/v1", api_key="lm-studio")

# Налаштування для потокового запиту
url = "http://192.168.0.105:1234/v1/responses"
headers = {
    "Content-Type": "application/json",
    "Accept": "text/event-stream"
}
payload = {
    "model": "mistralai/mistral-7b-instruct-v0.3",
    "input": "Write a 40-word story about a cat.",  # Довший промпт для помітного потокового виведення
    "stream": True
}

try:
    # Надсилаємо POST-запит із увімкненим потоком
    response = requests.post(url, headers=headers, json=payload, stream=True)
    response.raise_for_status()  # Перевірка на HTTP-помилки

    # Ініціалізація SSE-клієнта
    client_sse = sseclient.SSEClient(response)
    full_output = ""

    # Обробка подій
    for event in client_sse.events():
        print(f"Raw event: {event}")  # Виведення сирої події
        if event.data:
            try:
                # Спробуємо витягнути тип події з event.event
                event_type = event.event if event.event else None
                print(f"Parsed event type: {event_type}")

                # Спробуємо розпарсити event.data як JSON
                data = json.loads(event.data)
                print(f"Parsed data: {json.dumps(data, indent=2)}")

                # Перевірка типу події
                if event_type == "response.output_text.delta" or data.get("event") == "response.output_text.delta":
                    delta = data.get("delta", "")
                    full_output += delta
                    print(f"Delta: {delta}", end="", flush=True)
                elif event_type == "response.completed" or data.get("event") == "response.completed":
                    print("\nCompleted.")
                    break
            except json.JSONDecodeError as e:
                print(f"JSON Decode Error in event: {e}, Event data: {event.data}")
                # Ручна перевірка, якщо подія містить "response.output_text.delta"
                if "response.output_text.delta" in event.data:
                    try:
                        # Витягуємо delta з JSON-подібного рядка
                        delta = json.loads(event.data).get("delta", "")
                        full_output += delta
                        print(f"Delta: {delta}", end="", flush=True)
                    except json.JSONDecodeError:
                        print(f"Failed to extract delta from: {event.data}")
    print(f"\nFull output: {full_output}")

except requests.exceptions.RequestException as e:
    print(f"Error making request: {e}")
    if 'response' in locals():
        print(f"Server response status code: {response.status_code}")
except Exception as e:
    print(f"Unexpected error: {e}")
```
